```python
import daft
```

```python
df = daft.read_parquet("urls.parquet")
df.show()
```

```python
df = df.with_column("pdf_bytes", df["url"].url.download())
df.show()
```

```python
import pdfplumber
import io

@daft.udf(return_dtype=daft.DataType.string())
def extract_text(pdf_bytes_series):
    results = []
    for pdf_bytes in pdf_bytes_series.to_pylist():
        with pdfplumber.open(io.BytesIO(pdf_bytes)) as pdf:
            text = "\n".join([page.extract_text() for page in pdf.pages if page.extract_text()])
            results.append(text)
    return results
```

```python
df = df.with_column("extracted_text", extract_text(df["pdf_bytes"]))
df.show(3)
```

```python
df.write_parquet("extracted-text.parquet")
```

```python
daft.read_parquet("extracted-text.parquet").show()
```
